Este Laboratorio tiene como objetivo desarrollar un diagnostico experimental de paralelismo en pyhton, midiendo tiempos de ejecucion y analizando si las tareas pueden beneficiarse de computacion paralela.

# Librerias

```python
import time
import math
import random
import statistics
```

# Ejercicio 1

```python
def sum_cuadrados(n):
  return sum(i**2 for i in range(n))
```

```python
long = [100_000, 500_000, 1_000_000]

for i in long:
  start = time.perf_counter()
  result = sum_cuadrados(i)
  final = time.perf_counter()
  print(f"n={i:,} | resultado={result:,} | tiempo={final-start:.6f} s")
```

## Observacion

Se puede observar que el tiempo crece de forma aproximadamente proporcional a n, en cuanto n aumenta, el tiempo tambien aumenta. Es secuencial porque cada iteracion requiere de las iteraciones anteriores para actualizar.

# Ejercicio 2

```python
datos = list ( range (1 , 400_001 ) )
```

```python
def load(x):
  return math.sqrt(x) + math.log(x + 1)
```

## Secuencial


La version secuencial procesa cada elemento de la lista uno por uno, aplicando operacion matematica costosa(raiz cuadrada y logaritmo) a cada valor y acumulando el resultado.

```python
def sec(datos):
  result = 0.0
  result += sum(load(i) for i in datos)
  return result
```

## Por bloques

La version por bloques divide la lista en segmentos de 50000 elementos y procesa cada segmento por separado.

```python
def bloque(datos, t_bloque = 50000):
  result = 0.0
  for i in range(0, len(datos), t_bloque):
    trozo = datos[i : i + t_bloque]
    result += sum(load(j) for j in trozo)
  return result
```

## Comparacion

```python
# Secuencial
startS = time.perf_counter()
resultS = sec(datos)
finalS = time.perf_counter()
# Bloque
startB = time.perf_counter()
resultB = bloque(datos)
finalB = time.perf_counter()
print(f"Secuencial: resultado={resultS:,} | tiempo={finalS-startS:.6f} s \n Bloque: resultado={resultB:,} | tiempo={finalB-startB:.6f} s")
```

## Observacion

Ambas tardan tiempos similares al ser secuenciales, pero la version por bloque deja ver que cada bloque es independiente, dando una oportunidad de paralelizar ejecutando cada bloque en nucleos distintos.

# Ejercicio 3

```python
random.seed(42)
lotes = []

for i in range(8):
    lote = [random.uniform(0, 100) for j in range(120_000)]
    lotes.append(lote)

def resum(lote):
  return {
      "media": statistics.fmean(lote),
      "maximo": max(lote),
      "minimo": min(lote)
  }

start = time.perf_counter()
for i, lote in enumerate(lotes, start=1):
    r = resum(lote)
    print(f"Lote {i}: media={r['media']:.3f} | min={r['minimo']:.3f} | max={r['maximo']:.3f}")
final = time.perf_counter()
print(f"Tiempo total: {final-start:.6f} s")
```

## Observacion

los lotes son independientes entre si, y al no ser una cantidad de datos masiva no es optimo usar paralelizacion distribuida, por lo que lo mejor seria usar paralelizacion, ya que cada lote podria ejecutarse de forma independiente en cada nucleo de un procesador.

Para que la paralelizacion sea pertinente, los problemas deben tener tareas independientes entre si, es decir que no necesiten resultados de ejecuciones anteriores, y manejar una cantidad grande de datos.

### ¿Por qué no basta con que un programa “demore” para concluir que debe paralelizarse?.

No basta con que el problema demore para determinar si se debe paralelizar, como se pude ver en el ejercicio 1, la suma de los cuadrados es lenta pero cadda iteracion depende de la anterior, por lo que paralelizarla requiere coordinacion constante que podria ser mas costosa que el beneficio.
